```python
#career
games_career = games_df
# Group by player_id and calculate the sum for each of the specified columns
player_stats_career = games_career.groupby('player_id')[['goals', 'yellow_cards', 'red_cards', 'minutes_played', 'assists']].sum().reset_index()

# Rename columns for clarity if needed
player_stats_career.rename(columns={'goals': 'total_goals', 
                             'yellow_cards': 'total_yellow_cards', 
                             'red_cards': 'total_red_cards', 
                             'minutes_played': 'total_minutes_played',
                             'assists': 'total_assists'}, inplace=True)

player_stats_career
```

```python
games_2022 = games_df[games_df['year'] == 2022]

same_club_mask = games_2022['player_club_id'] == games_2022['player_current_club_id']

games_cur_club = games_2022[same_club_mask]

# Group by player_id and calculate the sum for each of the specified columns
player_stats_cur_club = games_cur_club.groupby('player_id')[['goals', 'yellow_cards', 'red_cards', 'minutes_played', 'assists']].sum().reset_index()

# Rename columns for clarity if needed
player_stats_cur_club.rename(columns={'goals': 'total_goals_cur_club', 
                             'yellow_cards': 'total_yellow_cards_cur_club', 
                             'red_cards': 'total_red_cards_cur_club', 
                             'minutes_played': 'total_minutes_played_cur_club',
                             'assists': 'total_assists_cur_club'}, inplace=True)

player_stats_cur_club
```

```python
def calculate_performance_score(row):
    row['min_per_game'] = row['minutes_played']/row['games_played']
    if row['position'] == 'Attack':
        return ((row['goals'] * 0.3) + (row['assists'] * 0.25) -
                (row['yellow_cards'] * 0.05) - (row['red_cards'] * 0.1) +
                (row['minutes_played'] * 0.1) + (row['games_played'] * 0.1) +
                (row['min_per_game'] * 0.1))
    elif row['position'] == 'Midfield':
        return ((row['goals'] * 0.15) + (row['assists'] * 0.25) -
                (row['yellow_cards'] * 0.15) - (row['red_cards'] * 0.15) +
                (row['minutes_played'] * 0.1) + (row['games_played'] * 0.1) +
                (row['min_per_game'] * 0.1))
    elif row['position'] == 'Defender':
        return ((row['goals'] * 0.05) + (row['assists'] * 0.1) -
                (row['yellow_cards'] * 0.25) - (row['red_cards'] * 0.3) +
                (row['minutes_played'] * 0.1) + (row['games_played'] * 0.1) +
                (row['min_per_game'] * 0.1))
    else:
        return None
```

```python
#short term
games_2022_21 = games_df[ (games_df['year'] == 2021) | (games_df['year'] == 2022) ]
# Group by player_id and calculate the sum for each of the specified columns
player_stats_2022_21 = games_2022_21.groupby('player_id')[['goals', 'yellow_cards', 'red_cards', 'minutes_played', 'assists']].sum().reset_index()

games_played_2022_21 = games_2022_21.groupby('player_id')['appearance_id'].count()

player_stats_2022_21 = pd.merge(player_stats_2022_21, games_played_2022_21, on = 'player_id', how = 'inner')

player_stats_2022_21 = pd.merge(player_stats_2022_21, players_df[['player_id', 'position']], how = 'inner')
player_stats_2022_21.rename(columns={'appearance_id':'games_played'}, inplace = True)


player_stats_2022_21['performance_2022'] = player_stats_2022_21.apply(calculate_performance_score, axis=1)

# Rename columns for clarity if needed
player_stats_2022_21.rename(columns={'goals': 'total_goals_2022', 
                             'yellow_cards': 'total_yellow_cards_2022', 
                             'red_cards': 'total_red_cards_2022', 
                             'minutes_played': 'total_minutes_played_2022',
                             'assists': 'total_assists_2022',
                             'games_played': 'games_played_2022'}, inplace=True)

player_stats_2022_21
```

```python
#features for declining/improving performance, 

#short term
games_2020_19 = games_df[games_df['year'] == 2021]
# Group by player_id and calculate the sum for each of the specified columns
player_stats_2020_19 = games_2020_19.groupby('player_id')[['goals', 'yellow_cards', 'red_cards', 'minutes_played', 'assists']].sum().reset_index()

games_played_2020_19 = games_2020_19.groupby('player_id')['appearance_id'].count()

player_stats_2020_19 = pd.merge(player_stats_2020_19, games_played_2020_19, on = 'player_id', how = 'inner')

player_stats_2020_19 = pd.merge(player_stats_2020_19, players_df[['player_id', 'position']], how = 'inner')
player_stats_2020_19.rename(columns={'appearance_id':'games_played'}, inplace = True)


player_stats_2020_19['performance_2020'] = player_stats_2020_19.apply(calculate_performance_score, axis=1)

# Rename columns for clarity if needed
player_stats_2020_19.rename(columns={'goals': 'total_goals_2020', 
                             'yellow_cards': 'total_yellow_cards_2020', 
                             'red_cards': 'total_red_cards_2020', 
                             'minutes_played': 'total_minutes_played_2020',
                             'assists': 'total_assists_2020',
                             'games_played': 'games_played_2020'}, inplace=True)

player_stats_2020_19
```

```python
# Define other weights and factors
EXPERIENCE_WEIGHT = 0.2
DECAY_2022 = 1.5
DECAY_2020 = 1.0

# Function to calculate the overall performance score
def calculate_performance_score(row):
    experiece_factor = (row['year'] / row['age']) * EXPERIENCE_WEIGHT
    time_decay_factor = (row['performance_2022'] * DECAY_2022) + (row['performance_2020'] * DECAY_2020)
    #market_value_factor = (row['highest_market_value_in_eur'] - row['market_value_in_eur'])
    #overall performance - long term
    return experiece_factor + time_decay_factor

# Apply the performance score calculation
final_merged_df['raw_performance_score'] = final_merged_df.apply(calculate_performance_score, axis=1)

# Normalize the scores to a 0-50 scale
max_score = final_merged_df['raw_performance_score'].max()
final_merged_df['normalised_performance_score'] = (final_merged_df['raw_performance_score'] / max_score) * 50
```
